```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/TODAS_CFOP_ARQUIVO.csv')
df
```

```python
# Crie uma nova coluna "Classificação" com base na coluna "TABELA DE CFOP"
df['Classificação'] = df['TABELA DE CFOP'].where(df['TABELA DE CFOP'].str.contains('Classificam-se'))

# Preencha as células vazias na coluna "Classificação" com os valores da linha anterior
df['Classificação'].fillna(method='ffill', inplace=True)

# Remova as linhas onde a coluna "Classificação" contém "Classificam-se"
df = df[df['TABELA DE CFOP'] != df['Classificação']]

# Redefina o índice do DataFrame
df.reset_index(drop=True, inplace=True)

df
```

```python
# Crie uma nova coluna "Classificação" com base na coluna "TABELA DE CFOP"
df['Classificação'] = df['TABELA DE CFOP'].where(df['TABELA DE CFOP'].str.contains('Classificam-se'))

# Preencha as células vazias na coluna "Classificação" com os valores da linha anterior
df['Classificação'].fillna(method='ffill', inplace=True)

# Remova as linhas onde a coluna "Classificação" contém "Classificam-se"
df = df[df['TABELA DE CFOP'] != df['Classificação']]

# Redefina o índice do DataFrame
df.reset_index(drop=True, inplace=True)

# Remova a segunda coluna "Classificação"
df.drop(columns=['Classificação'], inplace=True)
df
```

```python
# Dividir a coluna em duas com base no primeiro traço encontrado
df[['CFOP', 'Descrição']] = df['TABELA DE CFOP'].str.split('-', 1)

# Remover os espaços em branco no início e no final das strings
df['CFOP'] = df['CFOP'].str.strip()
df['Descrição'] = df['Descrição'].str.strip()

# Imprimir o DataFrame resultante
print(df)
```

```python
import pandas as pd

# Ler a primeira planilha e atribuir a um DataFrame chamado df_total
df_total = pd.read_excel('/content/drive/MyDrive/ArquivoConfiguradoECFOP.xlsx', sheet_name='Planilha1')

# Ler a segunda planilha e atribuir a um DataFrame chamado df_cfop_nova
df_cfop_nova = pd.read_excel('/content/drive/MyDrive/ArquivoConfiguradoECFOP.xlsx', sheet_name='Planilha2')

display(df_total)
display(df_cfop_nova)
```

```python

# Convertendo as colunas 'CFOP' e 'COD_CFOP' para o tipo de dados texto
df_total['CFOP'] = df_total['CFOP']*1000
#df_total['CFOP'] = df_total['CFOP'].astype(str)
#df_cfop_nova['COD_CFOP'] = df_cfop_nova['COD_CFOP'].astype(str)


# Vamos mesclar os dataframes df_total e df_cfop_nova com base na coluna "CFOP"
df_cfop_final = pd.merge(df_total, df_cfop_nova, left_on='CFOP', right_on='COD_CFOP')
df_cfop_final
```

```python
# Excluindo a coluna 'CFOP' do novo DataFrame
df_cfop_final = df_cfop_final.drop(columns=['CFOP'])

# Renomeando a coluna 'COD_CFOP' para 'CFOP'
df_cfop_final = df_cfop_final.rename(columns={'COD_CFOP': 'CFOP'})
df_cfop_final
```

```python
caminho_arquivo_excel = '/content/drive/MyDrive/NovasCFOP.xlsx'

# Salvando o DataFrame em um arquivo Excel
df_cfop_final.to_excel(caminho_arquivo_excel, index=False)
```
