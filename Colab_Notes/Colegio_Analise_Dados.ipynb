#Importar dados

--> Aqui importaremos as bibliotecas que serão utilizadas durante o desenvolvimento do algoritmo

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

--> Aqui vamos importar a biblioteca pandas e as tabelas que se encontram no documento xlsx (excel) no google drive

```python
# importar dados da tabela com nota dos alunos
caminho = '/content/drive/MyDrive/Excel/Colegio_notas_pasta.xlsx'
lista_tabela_notas = []
for i in range(9):
    tabela = pd.read_excel(caminho,sheet_name = i)
    lista_tabela_notas.append(tabela)
for tabela in lista_tabela_notas:
    print(tabela)
```

--> Aqui colocaremos o nome das matérias nas tabelas, utilizando um loop for e a função enumerate(). O nome será colocado na coluna que anteriormente era chamada de "Alunos"

```python
#Colocar o nome das materias nas tabelas
materias = ['portugues','matematica','biologia','historia','filosofia',
            'geografia','edu_fisica','fisica','quimica']
tabela_notas=[]
for i,tabela in enumerate(lista_tabela_notas):
    tabela = tabela.rename(columns={'Alunos':materias[i]})
    tabela_notas.append(tabela)
for tabela in tabela_notas:
    display(tabela)
```

--> Aqui criaremos uma coluna com média das notas de cada aluno para cada uma das matérias

```python
for tabela in tabela_notas:
    tabela['média'] = (tabela['prova_1'] + tabela['prova_2'] + tabela['prova_3'])/3
display(tabela_notas)
```

Aqui Criaremos uma tabela nova, que receberá os dados dos alunos com a maior nota em cada uma das matérias... Caso hajam dois alunos com a mesma nota, ambos devem aparecer na tabela, então devemos filtrar corretamente.

```python
tabela_melhores_desempenhos = pd.DataFrame()
nota = []
nome = []

n = -1
for tabela in tabela_notas:
    n += 1
    nota_index = tabela['média'].idxmax()
    nota.append(tabela['média'].max())
    nome.append(tabela.loc[nota_index,materias[n]])

tabela_melhores_desempenhos['Alunos'] = nome
tabela_melhores_desempenhos['notas'] = nota
tabela_melhores_desempenhos['matérias'] = materias
```

Veremos as maiores notas em cada matéria com base em um gráfico usando plotly express

```python
grafico_melhores_notas = px.line(x=tabela_melhores_desempenhos['matérias'],\
                                 y = tabela_melhores_desempenhos['notas'],\
                                 title = 'Melhores Notas: Linha',\
                                 template = 'plotly_dark',\
                                 labels = {'x':'matérias','y':'notas'})

grafico_melhores_notas.show()
```

O gráfico no formato line ficou muito feio e pouco entendível, então faremos um gráfico do tipo coluna, com o método px.bar()



```python
largura = 0.4
grafico_melhores_notas_coluna = px.bar(x=tabela_melhores_desempenhos['matérias'],\
                                 y = tabela_melhores_desempenhos['notas'],\
                                 title = 'Melhores Notas: Linha',\
                                 template = 'plotly_dark',\
                                 labels = {'x':'matérias','y':'notas'},\
                                 text = tabela_melhores_desempenhos['Alunos'])

grafico_melhores_notas_coluna.show()
```
