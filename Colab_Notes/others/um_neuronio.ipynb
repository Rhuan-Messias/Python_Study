```python
import numpy
```

```python
#               x1	    x2	    x3
matriz =   [[-0.3665, 0.0620, 5.9891],
            [-0.7842, 1.1267, 5.5912],
            [0.3012, 0.5611, 5.8234],
            [0.7757, 1.0648, 8.0677],
            [0.1570, 0.8028, 6.3040],
            [-0.7014, 1.0316, 3.6005],
            [0.3748, 0.1536, 6.1537],
            [-0.6920, 0.9404, 4.4058],
            [-1.3970, 0.7141, 4.9263],
            [-1.8842, -0.2805, 1.2548]]
pesos = numpy.random.rand(len(matriz))
bias = 1
```

```python
def ativacao(valor):
    if valor >= 0:
        return valor
    else:
        return 0
```

```python
v = [] # lista de resultados para cada exemplar
for vetor in matriz:
    forward_feeding = 0 
    for w,i in enumerate(vetor):
        forward_feeding = forward_feeding + i*pesos[w] + bias
    
    v.append(forward_feeding)

resultados_ativados = []
for i in v:
    resultados_ativados.append(ativacao(i))

resultados = []
for result in resultados_ativados:
    a = float(result)
    resultados.append(a)


print('Resultados depois da ativação:')
display(resultados)
```
