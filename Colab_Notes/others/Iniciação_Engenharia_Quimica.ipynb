cálculo da quantidade de corantes presente no adsorvente

V é o volume da solução no frasco e M é a massa seca de adsorvente

```python
import sympy as sp
q,v,c_i,c_t,M = sp.symbols("q,v,C_i,C_t,M")
quant_corante = sp.Eq(q,(v*(c_i-c_t)/M))
volume = 0
ci = 0
ct= 0
massa = 0
solve_quant_corante = sp.solve(quant_corante.subs({v:volume,
                                               c_i:ci,
                                               c_t:ct,
                                               M:massa
                                               }))[0]
solve_quant_corante = float(solve_quant_corante)
display(quant_corante)
print(f'quantidade de corante = {solve_quant_corante:.2f}')
```

```python
import sympy as sp
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
#Aqui as tabelas com resultados de comprimento de onda versus absorbância foram colocadas dentro dos respectivos dataframes
#para cada cor, vermelho e laranja, respectivamente.
df_vermelho = pd.read_excel('/content/drive/MyDrive/Iniciação Científica/Lacem/Vermelho_Laranja_Drimarem.xlsx',sheet_name='Vermelho')
df_laranja = pd.read_excel('/content/drive/MyDrive/Iniciação Científica/Lacem/Vermelho_Laranja_Drimarem.xlsx',sheet_name='Laranja')
display(df_vermelho)
display(df_laranja)
```

```python
# aqui, diversos gráficos serão traçados para obter informações úteis a partir dos dados coletados.

fig_1 = px.line(x=df_vermelho['Comprimento de Onda'],y=df_vermelho['Absorbância'],title='Vermelho Drimaren',labels={'x':'Comprimento de Onda',
                                                                                                                    'y':'Absorbância'},template='plotly_dark')
fig_2 = px.line(x=df_laranja['Comprimento de Onda'],y=df_laranja['Absorbância'],title='Laranja Drimaren',labels={'x':'Comprimento de Onda',
                                                                                                                    'y':'Absorbância'},template='plotly_dark')
fig_1.show()
fig_2.show()
```

```python
#aqui eu vou juntar os valores de absorbância do corante vermelho e laranja em
#uma mesma tabela, a fim de plotar um gráfico comparativo entre os valores
tabela_vermelho_laranja = pd.DataFrame()
tabela_vermelho_laranja['Comprimento de Onda'] = df_vermelho['Comprimento de Onda']
tabela_vermelho_laranja['Absorbância Vermelho'] = df_vermelho['Absorbância']
tabela_vermelho_laranja['Absorbância Laranja'] = df_laranja['Absorbância']
tabela_vermelho_laranja.index = tabela_vermelho_laranja['Comprimento de Onda']
tabela_vermelho_laranja = tabela_vermelho_laranja.drop('Comprimento de Onda', axis=1)
display(tabela_vermelho_laranja)
```

```python
#Aqui está o gráfico comparativo entre os resultados obtidos para os corantes vermelho e laranja
graf_comparativo = px.line(tabela_vermelho_laranja, template="plotly_white")
graf_comparativo.show()
```
