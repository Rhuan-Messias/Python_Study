```python
import pandas as pd 
```

It is important to read the raw .csv filein order to work
We'll only need the columns "movieId" and "title"

```python
ds_movies = pd.read_csv("https://raw.githubusercontent.com/krishnaik06/Recommendation_complete_tutorial/refs/heads/master/KNN%20Movie%20Recommendation/movies.csv")
ds_movies = ds_movies[["movieId","title"]]
ds_movies
```

```python
ds_rating = pd.read_csv("https://raw.githubusercontent.com/krishnaik06/Recommendation_complete_tutorial/refs/heads/master/KNN%20Movie%20Recommendation/ratings.csv")
ds_rating = ds_rating[["userId","movieId","rating"]]
ds_rating
```

Now we merge the two data_frame into one joining the movieId

```python
dataframe = pd.merge(ds_movies,ds_rating, on="movieId")
dataframe
```

Now, in order to pivot the dataframe, we'll put userId as the df index, and each title will be a column, having the rating as value. We'll create an 2x2 matrix where we can see which person gave each rating for each movie. fillna() will substitute the NaN for 0

```python
df_recommender = dataframe.pivot_table(index="userId", columns="title", values= "rating").fillna(0)
df_recommender
```

Now we find the K-Nearest Neighbours (KNN algorithm), so we can create the recommendations

```python
from sklearn.neighbors import NearestNeighbors

#hyper parameters is what adjust the learning model
#the common parameters are our data in the dataframe created after the pivot
modelo_knn = NearestNeighbors(metric='cosine')
modelo_knn.fit(df_recommender) #fit will find the nearest neighbours
```

df_recommender.iloc(idx_user) will return a Serie, an array with only 1 dimension
the model need a 2D array, so we use the array.reshape(1, -1) function. When it is (1, -1) we have a 2D in horizontal direction, while with (-1, 1) we have a 2D in vertical direction

```python
# let's recommend something to the user 230
# When we call the model, we need to know the K distance and who is the neighbours index
idx_user = 230 # the 230º user, not the real index

distance, neighbour_index = modelo_knn.kneighbors(df_recommender.iloc[idx_user].values.reshape(1, -1),
                                                  n_neighbors=3)
```

The first neighbour is the own person, that's why we can ignore the "0" proximity. 
The results below are 2D arrays, but if we use the "flatten()" function, we can make it 1D again.

```python
user = df_recommender.index[idx_user] # get the user index

for i in range(0, len(distance.flatten())):
    if i == 0:
        print("User: {0}\n".format(user))
    else:
        print("Neighbour {0} has distance {1}".format(df_recommender.index[neighbour_index.flatten()[i]],distance.flatten()[i]))
```

We can compare what the user and the nearest neightbours watched, and when the neighbour gave a good rating and the user didn't watch, we recommend this movie. It can be automatized in a loop for each neighbour. 
We can pass each condition when calling the ds_titles in parentesis and & --> ds_title = ds_title[(condition1) & (condition2) & (condition3)]

```python
ds_user = df_recommender.loc[user].to_frame() #make it a dataframe
ds_user
```

```python
next_neighbour = df_recommender.index[neighbour_index.flatten()[1]]
ds_neighbour = df_recommender.loc[next_neighbour].to_frame()
ds_neighbour
```

```python
ds_title = pd.merge(ds_user,ds_neighbour,on='title').sort_values(by=next_neighbour,ascending=False)
ds_title
```

```python
#next_neighbour and user are the real index(or name) for the nearest_neighbours_index list 
# we could recomend only those movies which the near neighbour gave > 3 of rating
ds_title_recommendation = ds_title[(ds_title[next_neighbour] > 3) & (ds_title[user]==0)].reset_index()
ds_title_recommendation
```
